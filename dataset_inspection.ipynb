```python
import pandas as pd

def print_board(row, board_dim):
    """
    Prints the board in a pretty format.
    """
    print("x" * (board_dim * 4 + 2))  # Top border

    for i in range(board_dim):
        # Leading spaces for offset
        

        print(" " * i, end="")
        print("o ", end="")  # Left border

        for j in range(board_dim):
            cell_value = row[f"cell{i}_{j}"]
            if cell_value == 1:
                print(" X ", end="")
            elif cell_value == -1:
                print(" O ", end="")
            else:
                print(" . ", end="")
            if j < board_dim - 1:
                print(" ", end="")  # Space between cells
        print(" o")  # Right border

    print(" " * board_dim + "x" * (board_dim * 4 + 2))  # Bottom border
    print(f"Winner: {'X' if row['winner'] == 1 else 'O'}")
    print()


def debug_dataset(file_path, board_dim):
    """
    Loads the dataset and prints each board with a key press.
    """
    df = pd.read_csv(file_path)

    for index, row in df.iterrows():
        print(f"Game {index + 1}")
        print_board(row, board_dim)
        command = input("Press Enter to show the next board...")
        if command == 'q':
            break

# Example usage
debug_dataset("hex_datasets/hex_4x4_10000_games_2_turns_before_win.csv", board_dim=4)
```
